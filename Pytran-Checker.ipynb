```python
import tkinter
from tkinter import *
from tkinter import ttk
from tkinter import Text
from textblob import TextBlob
from googletrans import Translator,LANGUAGES
import pyperclip
import webbrowser

root=Tk()
root.title("G-16 TTL PROJECT")
root.geometry("1200x400")
root.resizable(0,0)

# Load the image
background_image = PhotoImage(file="background.png")

# Create a label with the image as the background
background_label = Label(root, image=background_image)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

Center_place = Label(root,text="Pytran - checker",font="Elephant 20 bold",bg='#94cef3',fg="#0f1418")
Center_place.place(x=530,y=10)
center_icon= PhotoImage(file="translation.png")
tr_image = center_icon.subsample(10, 10)
center_label=Label(root,image=tr_image,bg='#94cef3')
center_label.place(x=480,y=10)


def check_spelling():
    word=Input_text.get()
    a=TextBlob(word)
    right=str(a.correct())

    cs=Label(root,text="Correct text is : ",font='ariel 12 bold',bg='#94cef3',fg='black')
    cs.place(x=20,y=200)
    spell.config(text=right)

    copy_button = Button(root, image=copy_image, bg='#94cef3', activebackground='#94cef3', bd=0, command=lambda: pyperclip.copy(text=right))
    copy_button.place(x=98, y=316)

Label(root,text='Input',font="Elephant 13",bg='#7282f3',fg="white").place(x=100,y=50)
Input_text=Entry(root,width=30,font='ariel 10',bg="white",border=2)
Input_text.place(x=20,y=80)

Label(root,text='Output',font="Elephant 13",bg='#94cef3',fg='#0f1418').place(x=92,y=150)

Label(root,text='Input',font="Elephant 13",bg='#94cef3',fg='#0f1418').place(x=1040,y=50)
Input_text_2=Entry(root,width=30,font='ariel 10',bg="white",border=2)
Input_text_2.place(x=960,y=80)

Label(root,text='Output',font="Elephant 13",bg='#94cef3',fg='#0f1418').place(x=1040,y=150)

spell_btn=Button(root,text="SpellCheck",font='arial 12 bold',pady=5,bg='#7282f3',activebackground='#3d34cb',fg="white",command=check_spelling)
spell_btn.place(x=335,y=80)

image_icon = PhotoImage(file="translation.png")
root.iconphoto(False,image_icon)

arrow_image = PhotoImage(file="transfer.png")
image_label=Label(root,image=arrow_image,bg='#94cef3')
image_label.place(x=536,y=125)

copy_image = PhotoImage(file="copy.png")

power_icon= PhotoImage(file="power-settings.png")
power_label=Label(root,image=power_icon,bg='#7282f3')
power_label.place(x=528,y=350)

power_label_2=Label(root,text="Powered By",font="Elephant 10",bg='#7282f3',fg="white")
power_label_2.place(x=562,y=350)

power_label_3=Label(root,text="2105045, 2105783, 2105793, 21052067 (KIIT, DU)",font="Elephant 7",bg='#7282f3',fg="white")
power_label_3.place(x=560,y=380)

kiit_icon= PhotoImage(file="KIIT-logo.png")
resized_image = kiit_icon.subsample(6, 6)
kiit_label=Label(root,image=resized_image,bg='#94cef3')
kiit_label.place(x=1120,y=-10)

language=list(LANGUAGES.values())

combo1=ttk.Combobox(root,values=language,font="ariel 12 bold",state="r")
combo1.place(x=498,y=80)
combo1.set("English")

combo2=ttk.Combobox(root,values=language,font="ariel 12 bold",state="r")
combo2.place(x=498,y=280)
combo2.set("Choose language")

spell=Label(root,font="ariel 12 bold",bg='#94cef3',fg='black')
spell.place(x=19,y=230)

Output_text=Text(root,font="ariel 12 bold",bg='#94cef3',fg='black',borderwidth=0, highlightthickness=0,height=2)
Output_text.place(x=970,y=230)

def Translate():
    translator=Translator()
    translated=translator.translate(text=Input_text_2.get(),dest=combo2.get())
    cs_2=Label(root,text="Translated text is : ",font='ariel 12 bold',bg='#94cef3',fg='black')
    cs_2.place(x=980,y=200)
    Output_text.delete(1.0,END)
    Output_text.insert(END,translated.text)
    
    copy_button = Button(root, image=copy_image, bg='#3a4cb2', activebackground='#3a4cb2', bd=0, command=lambda: pyperclip.copy(translated.text))
    copy_button.place(x=1040, y=316)

trans_btn=Button(root,text="Translate",font='arial 12 bold',pady=5,bg='#7282f3',activebackground='#3d34cb',fg="white",command=Translate)
trans_btn.place(x=765,y=80)

qa_icon=PhotoImage(file="feedback.png")
resize_qa=qa_icon.subsample(2,2)
qa_label=Label(root,image=resize_qa,bg='#94cef3')
qa_label.place(x=10,y=350)

def open_link(event):
    url = "https://forms.gle/1bjEu7UVCek4XcDM7"
    webbrowser.open_new(url)

qa_label2=Label(root,text="Have any queries?\nClick Here",font="Ariel 7",bg='#94cef3',cursor="hand2")
qa_label2.place(x=45,y=350)
qa_label2.bind("<Button-1>", open_link)

root.mainloop()
```
